```python
%%R

n.trials <- 100000

set.seed(17)
load_1.samples <- rnorm(n.trials, mean=8, sd=2)
dens_load_1 <- density(load_1.samples)

set.seed(21)
haul_1.samples <- rlnorm(n.trials, meanlog = 1.477, sdlog = 1)
dens_haul_1 <- density(haul_1.samples)

set.seed(23)
dump_1.samples <- rnorm(n.trials, mean=4, sd=1)
dens_dump_1 <- density(dump_1.samples)

par(mfrow = c(2, 2))
plot(dens_load_1, lwd = 2, col = "red",
     main = "Load,Haul,Dump Cycle Times 1", xlab = "",
     xlim = c(0, 30),
     ylim = c(0, max(dens_load_1$y, dens_haul_1$y, dens_dump_1$y)))
lines(dens_haul_1, col = "blue", lwd = 2)
lines(dens_dump_1, col = "yellow", lwd = 2)

cycle_total_1 <- load_1.samples + haul_1.samples + dump_1.samples
dens_cycle_total_1 <- density(cycle_total_1)
plot(dens_cycle_total_1, lwd=2, col = "green",
     main = "Total Cycle Time 1",
     xlim = c(0, 30),
     ylim = c(0, max(dens_load_1$y, dens_haul_1$y, dens_dump_1$y)))


set.seed(44)
load_2.samples <- rnorm(n.trials, mean=2, sd=1)
dens_load_2 <- density(load_2.samples)

set.seed(47)
haul_2.samples <- rlnorm(n.trials, meanlog = 1.653, sdlog = 0.954)
dens_haul_2 <- density(haul_2.samples)

set.seed(45)
dump_2.samples <- rnorm(n.trials, mean=2, sd=0.5)
dens_dump_2 <- density(dump_2.samples)

#par(mfrow = c(1, 2))
plot(dens_load_2, lwd = 2, col = "red",
     main = "Load, Haul, Dump Cycle Times 2", xlab = "",
     xlim = c(0, 30),
     ylim = c(0, max(dens_load_2$y, dens_haul_2$y, dens_dump_2$y)))
lines(dens_haul_2, col = "blue", lwd = 2)
lines(dens_dump_2, col = "yellow", lwd = 2)


cycle_total_2 <- load_2.samples + haul_2.samples + dump_2.samples
dens_cycle_total_2 <- density(cycle_total_2)
plot(dens_cycle_total_2, lwd=2, col = "green",
     main = "Total Cycle Time 2",
     xlim = c(0, 30),
     ylim = c(0, max(dens_load_2$y, dens_haul_2$y, dens_dump_2$y)))

```
