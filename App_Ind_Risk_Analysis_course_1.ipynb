# The Exponential Distribution

 The time-to-failure T of a component put into operation at time t = 0 has:

* probability density function
$$
f(t) = \left\{
    \begin{array}\\
      \lambda e^{-\lambda t} & \mbox{for } \ t > 0, & \lambda  > 0 \\
      0 & \mbox{otherwise }
    \end{array}
    \right.
$$
* cumulative density function
$$
F(t) = 1 - e^{-\lambda t}\\
$$ 
* survivor function aka survival probability function aka reliability function
$$
R(t) = e^{-\lambda t}\\
$$
* mean time-to-failure
$$
MTTF = \frac{1}{\lambda} \\
$$
